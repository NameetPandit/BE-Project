```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True' # very important
```

```python

import cv2
import easyocr

# Set your dataset path
dataset_path = "E:\\dataset\\images"
output_labels_path = os.path.join(os.path.dirname(dataset_path), "labels.txt")

# Initialize EasyOCR Reader
reader = easyocr.Reader(['en'])

# Open file to store labels
with open(output_labels_path, "w") as label_file:
    for img_name in os.listdir(dataset_path):
        img_path = os.path.join(dataset_path, img_name)
        
        # Read image
        img = cv2.imread(img_path)
        if img is None:
            print(f"❌ Skipping {img_name} (cannot read)")
            continue
        
        # Run OCR
        result = reader.readtext(img)
        
        # Extract detected text
        detected_text = "UNKNOWN"  # Default if no text found
        if result:
            detected_text = result[0][1]  # Get first detected text
        
        # Save label
        label_file.write(f"{img_name} {detected_text}\n")
        print(f"✅ {img_name} → {detected_text}")

print("🎉 All labels generated successfully!")
```

```python
import re

labels_path = "E:\\dataset\\labels.txt"

# Define a pattern for Indian license plates (basic format)
plate_pattern = re.compile(r'^[A-Z]{2}\d{1,2}[A-Z]{1,2}\d{4}$')

# Read labels and check validity
corrected_labels = []
with open(labels_path, "r") as file:
    for line in file:
        img_name, detected_text = line.strip().split(" ", 1)

        # Check if the detected text follows the pattern
        if not plate_pattern.match(detected_text):
            print(f"⚠️ Wrong Label Detected: {img_name} → {detected_text}")
            detected_text = "UNKNOWN"  # Mark as unknown for now

        corrected_labels.append(f"{img_name} {detected_text}")

# Save corrected labels
with open(labels_path, "w") as file:
    file.write("\n".join(corrected_labels))

print("✅ Labels checked and corrected (if needed).")
```
